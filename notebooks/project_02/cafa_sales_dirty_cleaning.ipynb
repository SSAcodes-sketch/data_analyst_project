```python
import pandas as pd
import numpy as np
df=pd.read_csv('cafe_sales_dirty.csv')
```

```python
df
```

```python
df_clear=df.dropna()
```

```python
df_clear
```

```python
print(f"Duplicate rows found: {df_clear.duplicated().sum()}")
```

```python
df_clear['Quantity'].dtype
```

```python
df_clear[['Quantity','Price Per Unit', 'Total Spent']] = df_clear[['Quantity','Price Per Unit', 'Total Spent']].apply(pd.to_numeric, errors='coerce')
df_clear_01=df_clear.dropna(subset=['Quantity','Price Per Unit', 'Total Spent'])
```

```python
df_clear_01['Quantity'] = df_clear_01['Quantity'].astype(int)
```

```python
valied_methods = ['Cash','Credit Card','Digital Wallet']
df_clear_01.loc[~df_clear_01['Payment Method'].isin(valied_methods), 'Payment Method'] = np.nan
df_clear_01=df_clear_01.dropna(subset=['Payment Method'])
```

```python
df_clear_01['Payment Method'].value_counts(dropna=False)
```

```python
valied_locations = ['In-store','Takeaway']
df_clear_01[~df_clear_01['Location'].isin(valied_locations)]= np.nan
df_clear_01=df_clear_01.dropna(subset=['Location'])
```

```python
df_clear_01
```

```python
# Shows a sample of the raw date strings
print(df_clear['Transaction Date'].unique()[:15])
```

```python
df_clear_01[df_clear_01['Transaction Date'].isin(['ERROR','UNKNOWN'])]=np.nan
df_clear_01=df_clear_01.dropna(subset=['Transaction Date'])
```

```python
df_clear_01['Transaction Date'] = pd.to_datetime(df_clear_01['Transaction Date'], errors='coerce')
```

```python
df_clear_01
```

```python
print(df_clear_01['Transaction ID'].dtype)
```

```python
# Look at rows that have duplicated Transaction IDs
duplicates = df_clear_01[df_clear_01.duplicated(subset=['Transaction ID'], keep=False)]
print(f"Number of duplicate rows found: {len(duplicates)}")
```

```python
df_clear_01['Transaction ID']=df_clear_01['Transaction ID'].astype(str).str.strip().str.upper()
```

```python
ID_length = len(df_clear_01['Transaction ID'][10])
```

```python
wrong_length = df_clear_01[df_clear_01['Transaction ID'].str.len() != ID_length]
print(f"Number of rows with incorrect ID length found: {len(wrong_length)}")
```

```python
bad_ids = df_clear_01[df_clear_01['Transaction ID'].str.contains('^[A-Z0-9]+$', regex=True)]
print(f"Number of rows with invalid characters in Transaction ID found: {len(bad_ids)}")
```

```python
print("=== 🌟 FINAL CLEAN DATASET PROFILE 🌟 ===")
print(f"Total Rows Remaining: {len(df_clear_01)}")
print(f"Total Columns: {df_clear_01.shape[1]}")
print("\n--- 📊 Data Types & Missing Values ---")
print(df_clear_01.info())

print("\n--- 🧼 Unique Value Verification ---")
print("Payment Methods found:", df_clear_01['Payment Method'].unique())
print("Any Duplicate Transaction IDs?:", df_clear_01.duplicated(subset=['Transaction ID']).any())
```

```python
# Save the cleaned DataFrame to a new CSV file
# index=False prevents Pandas from adding an annoying extra column of row numbers
df_clear_01.to_csv('cleaned_transactions_data.csv', index=False)

print("🚀 Clean CSV successfully saved as 'cleaned_transactions_data.csv'!")
```
