```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('laptop_data_cleaned.csv')
```

```python
df
```

```python
fig,ax=plt.subplots(2,1,figsize=(10,10))

sns.countplot(
    data=df,
    x='Cpubrand',
    ax=ax[0],
    order=df['Cpubrand'].value_counts().index,
    palette='Greens_r'
)
ax[0].set_title('Total Laptop Count by CPU Brand', fontsize=14, fontweight='bold', pad=15)
ax[0].set_xlabel('CPU Brand', fontsize=12)
ax[0].set_ylabel('Number of Laptops (Frequency)', fontsize=12)
ax[0].tick_params(axis='x', rotation=30)
top_6_companies = df['Company'].value_counts().index[:6]
sns.countplot(
    data=df,
    x='Company',
    ax=ax[1],
    palette='Greens_r',
    order=top_6_companies
)

ax[1].set_title('Total Laptop Count by Company', fontsize=14, fontweight='bold', pad=15)
ax[1].set_xlabel('Company', fontsize=12)
ax[1].set_ylabel('Number of Laptops (Frequency)', fontsize=12)
ax[1].tick_params(axis='x', rotation=30)
ax[1].set_ylim(0,500)

fig.tight_layout()
plt.show()
```

```python
fig,ax=plt.subplots(2,1,figsize=(10,10))
sns.boxplot(
    data=df[df['Company'].isin(top_6_companies)],
    x='Company',
    y='Price',
    order=top_6_companies,
    ax=ax[0],
    palette='Set2'
)
ax[0].axhline(y=100000, color='red', linestyle='--', linewidth=1.5, label='₹100,000 Threshold')
ax[0].set_title('Laptop Price Distribution Across Top 6 Companies', fontsize=14, fontweight='bold', pad=15)
ax[0].set_xlabel('Company', fontsize=12)
ax[0].set_ylabel('Price (₹ INR)', fontsize=12)
ax[0].tick_params(axis='x', rotation=30)
ax[0].set_ylim(0, 300000)
ax[0].ticklabel_format(style='plain', axis='y') 
ax[0].legend()

sns.boxplot(
    data=df,
    x='Cpubrand',
    y='Price',
    order=df['Cpubrand'].value_counts().index,
    ax=ax[1],
    palette='Set2'
)
ax[1].axhline(y=100000, color='red', linestyle='--', linewidth=1.5, label='₹100,000 Threshold')
ax[1].set_title('Laptop Price Distribution Across CPU Brands', fontsize=14, fontweight='bold', pad=15)
ax[1].set_xlabel('CPU Brand', fontsize=12)
ax[1].set_ylabel('Price (₹ INR)', fontsize=12)
ax[1].tick_params(axis='x', rotation=30)
ax[1].set_ylim(0, 300000)
ax[1].ticklabel_format(style='plain', axis='y') 
ax[1].legend()

fig.tight_layout()
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.scatterplot(
    data=df,
    x='Ram',
    y='Price',
    alpha=0.4,          # Makes dots slightly transparent so you can see where they overlap/cluster
    color='darkgreen',
    s=80,
    ax=ax
)
unique_ram = sorted(df['Ram'].unique())
ax.set_xticks(unique_ram)
ax.set_xticklabels([f"{int(x)}GB" for x in unique_ram])
ax.tick_params(axis='x', rotation=45)
ax.set_title('Relationship Between RAM Capacity and Laptop Price', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('RAM (GB)', fontsize=12)
ax.set_ylabel('Price (₹ INR)', fontsize=12)
ax.ticklabel_format(style='plain', axis='y')
ax.set_xlim(0,70)

fig.tight_layout()
plt.show()
```
