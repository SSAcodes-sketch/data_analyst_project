```python
import pandas as pd
df=pd.read_csv('../data/laptopData.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_clear=df.dropna().copy()
```

```python
print(f"Duplicate rows found: {df_clear.duplicated().sum()}")
```

```python
df_clear.drop_duplicates(inplace=True)
```

```python
df_clear.info()
```

```python
df_clear['Inches']=df_clear['Inches'].astype(str).str.replace('Inches','',case=False)
df_clear['Inches']=df_clear['Inches'].astype(str).str.replace('"','').str.strip()
df_clear['Inches']=pd.to_numeric(df_clear['Inches'], errors='coerce')
```

```python
df_clear['Ram']=df_clear['Ram'].astype(str).str.replace('GB','',case=False)
df_clear['Ram']=pd.to_numeric(df_clear['Ram'], errors='coerce').astype('Int64')
```

```python
df_clear['Weight']=df_clear['Weight'].astype(str).str.replace('kg','',case=False)
df_clear['Weight']=pd.to_numeric(df_clear['Weight'], errors='coerce')
```

```python
df_clear = df_clear[df_clear['Weight'] > 0.5]
```

```python
df_clear[['Inches', 'Ram', 'Weight']].describe()
```

```python
df_clear['Cpu'].value_counts()
```

```python
df_clear['Cpubrand']=df_clear['Cpu'].apply(lambda x:' '.join(x.split()[0:3]))
```

```python
df_clear
```

```python
df_clear.Cpubrand.value_counts()
```

```python
cpu_counts = df_clear['Cpubrand'].value_counts()
minor_cpus = cpu_counts[cpu_counts < 15].index
minor_cpus
df_clear['Cpubrand'] = df_clear['Cpubrand'].replace(minor_cpus, 'Other')
```

```python
minor_cpus
```

```python
df_clear['Cpubrand'].value_counts()
```

```python
# 1. Save a clean, lightweight CSV for backup and model training
df_clear.to_csv('laptop_data_cleaned.csv', index=False)

# 2. Excel structure setup (Optional but excellent for organization)
with pd.ExcelWriter('laptop_data_workspace.xlsx', engine='openpyxl') as writer:
    df_clear.to_excel(writer, sheet_name='Cleaned_Data', index=False)
    
print("Successfully saved clean data files to workspace! 🎉")
```
